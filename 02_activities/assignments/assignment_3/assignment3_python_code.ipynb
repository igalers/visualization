```python
import json
file_path = 'C:/Users/igal_/Downloads/cycling-network - 4326.csv'
df = pd.read_csv(file_path)

# Function to extract (lat, lon) from the 'geometry' column
def extract_coordinates(geometry):
    geo_json = json.loads(geometry)  # Convert GeoJSON string to dictionary
    if geo_json['type'] == 'MultiLineString':
        # Extract the first coordinate pair (lon, lat) and reverse to (lat, lon)
        coordinates = geo_json['coordinates'][0][0]
        return coordinates[1], coordinates[0]  # Return (lat, lon)
    return None, None

# Apply the function and create new columns for lat and lon
df[['lat', 'lon']] = df['geometry'].apply(extract_coordinates).apply(pd.Series)

# Show a sample of the DataFrame to verify
print(df[['geometry', 'lat', 'lon']].head())
```

```python
fig = px.scatter_mapbox(df, 
                        title = "Cycling Network in the GTA",
                        lat="lat", 
                        lon="lon", 
                        hover_name="STREET_NAME",
                        zoom=9.5,  
                        mapbox_style="open-street-map")
fig.update_layout(
    title_x=0.5,
    )

fig.show()
```
