```python
#import required libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import PIL 
import requests
import scipy
```

```python
#add seed for reproducibility
np.random.seed(613)
x= np.arange(50)
y1= np.random.randint(0,100,50)
y2= np.random.randint(0,100,50)
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x, y1, label= 'Person 1')
ax.plot(x, y2, label='Person 2')
ax.legend(loc= 'lower right', #choose legend location
          frameon=True, #add frame around the legend
          fontsize = 12,
          ncol = 2, #change number of columns
          shadow = True) #add drop shadow
```

```python
x1 = np.arange(50)
y1 = np.random.randint(0,75,50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([110, 180, 240, 99, 220])

#create two plots next to each other
fig, (ax1, ax2) = plt.subplots(ncols=2,
                            nrows=1,
                            figsize = (7,3)
                            )

ax1.scatter(x1,y1,
            marker = "+",
            color = 'green'
            )
ax2.bar(x2,y2,
        facecolor = 'blue'
        )


                        
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1',  'ax3'], 
                                ['ax2', 'ax3']],
                                figsize = (7,4),
                                layout = 'constrained')

someaxes ["ax1"].scatter(x1,y1)
someaxes ["ax2"].bar(x2, y2)
someaxes ['ax3'].plot(x1, y1)

someaxes ["ax1"].set_xlabel('A Big Label', fontsize = 18)
someaxes ["ax2"].set_xlabel('Another Label', fontsize = 18)
someaxes ["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize = 18)
```

```python
x = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 50, 120])

fig, ax = plt.subplots(figsize= (7,3))

ax.bar(x,y1, color = 'indigo')
ax.plot(x, y2, color = 'red')

```

```python
#calculate std deviation
y2_sd = np.std(y2)

fig, ax = plt.subplots(figsize= (7,3))
ax.plot(x, y2, color = 'red')

#add error bar to the visualization
ax.errorbar(x, #our x values
            y2, #our y values
            yerr=y2_sd, #add error bars
            fmt="None",
             ecolor="indigo",
            elinewidth= 4,
             capsize=6,
              capthick=4,
              errorevery=2) #assigns error bar every interval 
```
