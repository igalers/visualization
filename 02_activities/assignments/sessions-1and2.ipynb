```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#add a seed to ensure reproducibility
np.random.seed(613)

#pick random numbers up to 50 on x-axis
x = np.arange(50)

#pick random integers between 0 and 100
y = np.random.randint(0,100,50)

#
fig, ax = plt.subplots(figsize = (5,3))
#ax.scatter(x,y) - to draw a scatterplot
#ax.bar(x,y) - to draw a bar graph
#ax.plot(x,y) - to draw a plot

#draw a histogram
ax.hist(y)
```

```python
#assign font type and size
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}

fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)

#add title and labels to your graph
ax.set_title('Total growth over time', fontdict=font1, loc = 'right')
ax.set_ylabel('Total growth', fontdict=font2)
ax.set_xlabel('Years since start', fontdict=font2)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (5,3))
#change marker type and color
ax.scatter(x,y,
           marker = '*',
           color = 'indigo',
           )
```

```python
fig, ax = plt.subplots(figsize = (5,3))
#change line width, color and style 
#change marker size, edge color and face color
ax.plot(x,y,
           marker = '*',
           color = '#7425b9',
           linestyle = '--',
           linewidth = 2,
           markersize = 12,
           markeredgecolor='#fa0359',
           markerfacecolor='black'
           )
#edit grid lines
ax.grid(axis='y',
         linestyle = '-',
         linewidth = 0.5,
         color = 'green'
         )
```

```python
#import relevant package
import seaborn as sns
```

```python
#create a sample data frame
df = pd.DataFrame(
    np.random.random((10,10)),
    columns=["a","b","c","d","e","f","g","h","i","j"]
)
```

```python
# plot a heatmap with annotation
sns.heatmap(df, annot=True, annot_kws={"size": 7})
plt.show()
```

```python
# plot a heatmap with custom grid lines
sns.heatmap(df, linewidths=2, linecolor='yellow')
plt.show()
```

```python
#remove labels
sns.heatmap(df, yticklabels=False)
plt.show()
```

```python
# plot a heatmap
sns.heatmap(df, cbar=False) 
plt.show()
```
